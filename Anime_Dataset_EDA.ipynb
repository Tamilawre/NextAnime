```python
# Parse the stringed list into an actual Python list
def extract_genres(genre_string):
    try:
        # Remove any leading/trailing whitespace and evaluate the string as a list
        genres = ast.literal_eval(genre_string.strip())
        # Ensure we have a flat list of genres
        return [g.strip() for g in genres if isinstance(g, str)]
    except (ValueError, SyntaxError):
        # If evaluation fails, return an empty list
        return []

# Convert the genre column to a list
genre_lists = anime['genre'].tolist()

# Extract genres from each stringed list
all_genres = [genre for genres in genre_lists for genre in extract_genres(genres)]

# Get unique genres
genres = sorted(set(all_genres))
```

```python
genres
```

```python
anime_genres = dict()
for genre in genres:
    anime_genres[genre] = []
```

```python
# One hot encode the genres
for genre_list in genre_lists:
    for genre in genres:
        if genre in genre_list:
            anime_genres[genre].append(1)
        else:
            anime_genres[genre].append(0)
```

```python
embeddings = pd.DataFrame(anime_genres)
embeddings.set_index(anime["uid"], inplace=True)
embeddings
```
